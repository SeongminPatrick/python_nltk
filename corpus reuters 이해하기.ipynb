## 조동사에 대해 얼마나 사용하는 지를 확인 

```python
import nltk
from nltk.corpus import reuters

news_text = reuters.words(categories='housing')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print( m + ':', fdist[m])
```

## coditionalFregDist 클래스에 대해 이해하기 

```python
cfd = nltk.ConditionalFreqDist( (genre, word) for genre in reuters.categories() for word in reuters.words(categories=genre))
genres = ['groundnut', 'groundnut-oil', 'heat', 'hog', 'housing', 'income', 'instal-debt'] 
modals = ['can', 'could', 'may', 'might', 'must', 'will'] 

cfd.tabulate(conditions=genres, samples=modals)
```

```python
cfd.conditions()
```
