## 품사별로 추출해서 보기 

### 말뭉치 중에 특정 범주로 가져와서 품사 중에 동사에 해당하는 것만 추출해서 빈도수에  맞춰 처리 하기 

```python
wt = []
brown_a = nltk.corpus.brown.tagged_words(categories="news")
for (word, tag) in brown_a:
    if tag[:2] == 'VB' :
        wt.append(word + '/' + tag)
fd=nltk.FreqDist(wt)


print(sorted(fd)[:20])

print(list(fd.items())[:10])
```

### 명사만 추출 

```python
nouns = []
for (w, t) in brown_a:
           if t[:2] == 'NN':
                nouns.append(w + '/' + t)
nfd = nltk.FreqDist(nouns)

print(sorted(nfd)[:20])
```

```python
def brown_tagged_sents(genre):
    """Returns the tagged sentences of the given Brown category."""
    return nltk.corpus.brown.tagged_sents(categories=genre)
```

# 연설문을 가져오기 

```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
state_union.fileids()
```

```python
train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
```

## 문장을 분리하는 기준을 넣어서 처리

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)

print(custom_sent_tokenizer)
```

```python
help(PunktSentenceTokenizer)
```

##  실제 문장 단위로 토큰을 분리 

```python
tokenized = custom_sent_tokenizer.tokenize(sample_text)

print(tokenized[:5])
```

## 단어로 토큰화하고 그 품사를 달기 

```python
def process_content():
    try:
        for i in tokenized[:5]:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)

    except Exception as e:
        print(str(e))


process_content()
```
